```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

classes = ["idle", "run", "walk"]
```

```python
# cm = confusion_matrix(y_true, y_pred)
cm = np.array([[0.995, 0.00, 0.005], [0, 0.9965, 0.0035], [0.0083, 0.0053, 0.9864]])*100
g = sns.heatmap(cm, annot=True, fmt='.2f', cmap="Blues",
                xticklabels=classes, yticklabels=classes)
g.set_xticklabels(g.get_xticklabels(), ha='center', rotation=0)
g.set_yticklabels(g.get_yticklabels(), rotation=0)
plt.ylabel("True Label")
plt.xlabel("Predicted Label")
```

```python
import json

with open("/Users/dcongtinh/Workspace/endless-runner/results/20210530_154012_LSTM_Action_Tanh/k6/history.json") as f:
    history = f.read()
history = eval(json.loads(history))
```

```python
def save_fig(epoch_arr, name, train, val):
    plt.plot(epoch_arr, train, 'g-o', label='Training ' + name, markersize=3)
    plt.plot(epoch_arr, val, 'r-o', label='Validation ' + name, markersize=3)
    plt.title(name + ' Figures')
    plt.legend()
    plt.xlabel("Epochs")
    plt.ylabel(name)
    # plt.savefig(filepath, facecolor='w', transparent=False)
    # plt.clf()
```

```python
epoch_arr = [i for i in range(50)]
train_acc = history['accuracy']
train_auc = history['auc']
train_loss = history['loss']

val_acc = history['val_accuracy']
val_auc = history['val_auc']
val_loss = history['val_loss']

save_fig(epoch_arr, 'Accuracy', train_acc, val_acc)
```

```python
save_fig(epoch_arr, 'AUC', train_auc, val_auc)
```

```python
save_fig(epoch_arr, 'Loss', train_loss, val_loss)
```
